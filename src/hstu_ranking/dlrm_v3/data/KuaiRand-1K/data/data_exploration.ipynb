```python
import pandas as pd
```

```python
df = pd.read_csv('processed_seqs.csv')
df.head(2)
```

```python
# map action_weights to binary values: if action_weights > 0, set to 1; if action_weights = 0, set to 0
def action_weights_to_binary(action_weights_str):
    action_weights = eval(action_weights_str)
    action_weights = [1 if weight > 0 else 0 for weight in action_weights]
    return str(action_weights)

df['action_weights'] = df['action_weights'].apply(action_weights_to_binary)

df.head(2)
```

```python
df.to_csv('processed_seqs_binary.csv', index=False)
```
